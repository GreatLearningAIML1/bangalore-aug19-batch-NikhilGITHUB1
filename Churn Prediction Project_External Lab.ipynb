```python
import pandas as pd
import numpy as np
import keras 
import matplotlib.pyplot as plt

```

```python
bank_df=pd.read_csv("bank.csv")
```

```python
bank_df.head()
```

```python
bank_df.shape
```

```python
#dropping the CustomerId and the Row Number
bank_df.drop(['RowNumber','CustomerId','Surname'], axis=1, inplace=True)
bank_df.head()
```

```python
# Create Training and Test sets

from sklearn.model_selection import train_test_split

y = bank_df['Exited']
X = bank_df.drop('Exited',axis=1)

X = pd.get_dummies(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Create the layers
classifier = Sequential()

# Adding the first hidden layer
classifier.add(Dense(units=6,
                     kernel_initializer='uniform',
                     activation='relu',
                     input_dim=13))

# Adding the second hidden layer
classifier.add(Dense(units=6,
                     kernel_initializer='uniform',
                     activation='relu'))  # don't need to specify the input_dim for subsequent layers

# Add the final output layer
classifier.add(Dense(units=1,   # we have a binary outcome Yes/No
                     kernel_initializer = 'uniform',
                     activation = 'sigmoid'))  # for more than 2 categories use 'softmax'
```

```python

# Compile the ANN
classifier.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
```

```python
# Fit the ANN to the data
classifier.fit(X_train, y_train, batch_size=10, epochs=100)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

y_pred = classifier.predict(X_test)
y_pred = y_pred > 0.5

print ("Accuracy on test set : ", accuracy_score(y_test, y_pred))

print ("\nConfusion Matrix : \n", confusion_matrix(y_test, y_pred))
```
