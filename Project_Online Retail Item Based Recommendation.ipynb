```python
import pandas as pd
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
import numpy as np
```

```python
retail_df=pd.read_excel("Online Retail.xlsx",sheet_name="Online Retail")
```

```python
retail_df.head()
```

```python
retail_df['Description'] = retail_df['Description'].str.strip()
retail_df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
retail_df['InvoiceNo'] = retail_df['InvoiceNo'].astype('str')
retail_df = retail_df[~retail_df['InvoiceNo'].str.contains('C')]
```

```python
retail_df.isnull().sum()
```

```python
median_value=retail_df['CustomerID'].median()
retail_df['CustomerID']=retail_df['CustomerID'].fillna(median_value)
```

```python
retail_df.shape
```

```python
pivot_ratings = retail_df.pivot_table(values = 'Quantity', index = 'StockCode', columns = 'Country')
```

```python
pivot_ratings
```

```python
pivot_ratings.fillna(0, inplace = True)
```

```python
pivot_ratings
```

```python
product_sparse = csr_matrix(pivot_ratings.values)
```

```python
query_index = np.random.choice(pivot_ratings.shape[0])
```

```python
query_index
```

```python
model_retail = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_retail.fit(product_sparse)

model_retail
```

```python
distances, indices = model_retail.kneighbors(pivot_ratings.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
print(distances.flatten())
print(indices.flatten())
```

```python
pivot_ratings.iloc[query_index,:].values.reshape(1, -1).shape
```

```python
indices.ndim
```

```python
pivot_ratings.head()
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(pivot_ratings.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, pivot_ratings.index[indices.flatten()[i]], distances.flatten()[i]))
```
